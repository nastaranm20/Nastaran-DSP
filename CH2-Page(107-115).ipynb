# فصل دوم-صفحه 107
## پاسخ پله واحد سیستم خطی تغییرناپذیر با زمان
<div dir="rtl">
تا اینجا دیدیم که نمایش سیستمهای LTI برحسب پاسخ ضربه آنها این امکان میدهد که عبارتهای بسیار صریحی برای خواص سیستم بدست آوریم. مشخصا بخاطر اینکه $h[n]$ یا $h(t)$ رفتار سیستم LTI را بطور کامل مشخص میکند، توانستیم خواصی مثل پایداری و علی بودن را به ویژگیهای پاسخ ضربه مرتبط کنیم.
سیگنال دیگری هم وجود دارد که آنهم غالبا برای توصیف رفتار سیستمهای LTI بکار میرود: پاسخ پله واحد $s[n]$ یا $s(t)$، که پاسخ سیستم به ازای $u[n]$ یا $u(t)$ است. مراجعه به پاسخ پله در بعضی اوقات مفید است، بنابراین یافتن رابطه آن با پاسخ ضربه به زحمتش می‌ارزد. باتوجه به نمایش جمع کانولوشن پایخ پله یک سیستم گسسته در زمان LTI کانولوشن پله واحد :و پاسخ ضربه است؛ یعنی
$$ s[n]=u[n]*h[n] $$
ولی با توجه به خاصیت جابجایی کانولوشن داریم $s[n]=h[n]*u[n]$، بنابراین $s[n]$ را میتوان پاسخ یک سیستم گسسته در زمان LTI با پاسخ ضربه $u[n]$، به ورودی $h[n]$ دانست. در مثال 2-12 دیدیم که $u[n]$ پاسخ ضربه انباره است. پس:
$$ s[n]=\sum_{k=-\infty}^{n}h[k]\;\;\;\;\;\;\; (91-2) $$
از این معادله و مثال 2-12 چنین برمی‌آید که $h[n]$ را میتوان با استفاده از رابطه زیر از $s[n]$ بازیابی کرد:
$$ h[n] = s[n-1]-s[n]\;\;\;\;\;\;\; (92-2) $$
یعنی پاسخ پله سیستمهای LTI گسسته در زمان جمع انباره‌ای پاسخ ضربه آنهاست (معادله 2-91). برعکس، پاسخ ضربه سیستمهای LTI گسسته در زمان تفاضل اول پاسخ پله آنهاست (معادله 2-92).
به نحوی مشابه، در حالت پیوسته در زمان پاسخ پله سیستمهای LTI دارای پاسخ ضربه $h(t)$ عبارت است از $s(t)=u(t)*h(t)$، که با پاسخ یک انتگرالگیر (که پاسخ ضربه‌اش $u(t)$ است) به ورودی $h(t)$ برابر است. یعنی پاسخ پله سیستمهای LTI پیوسته در زمان انتگرال انباره‌ای پاسخ ضربه آنهاست، یا:
$$ s(t)=\int_{-\infty}^{t}h(\tau)d\tau\;\;\;\;\;\;\;\;\;(93-2) $$
باتوجه به معادله 2-93، پاسخ ضربه مشتق پاسخ پله است، یا:
$$ h(t)=\frac{ds(t)}{dt}=s'(t)\;\;\;\;\;\;\;\;(94-2) $$
پس در هردوحالت پیوسته و گسسته در زمان، پاسخ پله را نیز میتوان برای مشخص کردن یک سیستم LTIبکار برد، زیرا پاسخ ضربه را میتوان از آن بدست آورد. در مسئله 2-45 عباراتی مشابه جمع و انتگرال کانولوشن، برای نمایش سیستم LTIبر حسب پاسخ پله  واحد بدست میاوریم.

# فصل دوم-صفحه 108
## سیستمهای علی توصیف شده با معادلات دیفرانسیل و تفاضلی
<div dir="rtl">
یک دسته بسیار مهم از سیستمهای پیوسته در زمان سیستمهایی هستند که ورودی و خروجی‌شان توسط یک معادله دیفرانسیل خطی با ضرائب ثابت بهم مربوط میشود. این معادلات در توصیف سیستمها و پدیده‌های فیزیکی بسیار متنوعی به کار میایند. برای نمونه در فصل 1 نشان دادیم که پاسخ مدار RC شکل 1-1 و حرکت خودرو در اثر ورودیهای شتاب‌دهنده و نیروهای اصطکاکی نشان داده شده در شکل 1-2 را میتوان با معادلات دیفرانسیل خطی دارای ضرائب ثابت توصیف کرد. در توصیف سیستمهای مکانیکی دارای نیروهای برگرداننده و میراساز، واکنشهای شیمیایی و بسیاری زمینه‌های دیگر نیز با این نوع معادلات دیفرانسیل روبرو میشویم.
به نحوی مشابه، دسته مهمی از سیستمهای گسسته در زمان هم سیستمهایی هستند که ورودی و خروجی‌شان با معادلات تفاضل خطی با ضرائب ثابت بهم مربوط میشوند. اینگونه معادلات برای توصیف رفتار ترکیبی فرایندهای متفاوتی بکار میروند. مثلا در مثال 1-10 دیدیم که چگونه توصیف انباشته شدن حساب بانکی به یک معادله تفاضلی انجامید، و در مثال 1-11 چگونگی کاربرد معادلات تفاضلی در شبیه‌سازی دیجیتالی سیستمهای پیوسته در زمان توصیف شده با معادلات دیفرانسیل را دیدیم.
معادلات تفاضلی در مشخص کردن سیستمهای گسسته در زمان طراحی شده برای انجام کارهای مشخص بر روی سیگنال ورودی نیز ظاهر میشوند. مثلا سیستمی که تفاضل بین ورودیهای متوالی را حساب میکند، معادله 1-99، و سیستم توصیف شده با معادله 1-104 که مقدار متوسط ورودی در یک فاصله را محاسبه میکند با معادلات تفاضلی توصیف میشوند.
در این کتاب در موارد متعددی سیستمهای توصیف شده با معادلات دیفرانسیل یا تفاضلی خطی با ضرائب ثابت را در نظر میگیریم و بررسی میکنیم. در این بخش ابتدا نگاهی به این سیستمها می‌اندازیم تا ایده‌های اصلی حل معادلات دیفرانسیل و تفاضلی را معرفی کرده، برخی خواص سیستمهای توصیه شده با این معادلات را آشکار و بررسی میکنیم. در فصلهای بعدی ابزارهای جدیدی برای تحلیل سیگنالها و سیستمها پی‌ریزی میکنیم؛ این ابزارها هم توانایی‌مان در تحلیل سیستمهای توصیف شده با این معادلات را افزایش میدهند و هم درک بهتری از مشخصات و رفتارشان بدست میدهند.

# فصل دوم-صفحه 109
## معادلات دیفرانسیل خطی با ضرائب ثابت
<div dir="rtl">
برای شروع معرفی مفاهیم مهم مربوط به سیستمهای توصیف شده با معادلات دیفرانسیل خطی با ضرائب ثابت، معادله دیفرانسیل مرتبه اول 1-85 را در نظر میگیریم:
$$ \frac{dy(t)}{dt}+2y(t)=x(t)\;\;\;\;\;\;\;\;(95-2) $$
که در آن $y(t)$ خروجی و $x(t)$ ورودی سیستم است. مقایسه معادله 2-95 و معادله دیفرانسیل 1-84 مربوط به سرعت خودروی در معرض نیروهای اصطکاکی، نشان میدهد که اگر $y(t)$ را سرعت $v(t)$ بگیریم و $x(t)$ را نیروی اعمال شده به خودرو، و اگر پارامترهای معادله 1-84 به نحوی باشند که $b/m=2$ و $1/m=1$، این معادله دقیقا به آن سیستم مربوط میشود.
یک نکته‌ی بسیار مهم در مورد معادلات دیفرانسیلی مثل معادله‌ی 2-95 این است که این معادلات سیستم را به طور ضمنی مشخص میکنند. یعنی رابطه‌ی بین خروجی و ورودی را بیان میکنند، نه اینکه خروجی را به طور صریح بصورت تابعی از ورودی بدست دهند. برای یافتن عبارت صریح باید معادله دیفرانسیل را حل کنیم. برای یافتن جواب اطلاعاتی بیش از آنچه معادله دیفرانسیل دربر دارد، نیاز داریم. مثلا برای تعیین سرعت خودرو در انتهای یک فاصله 10ثانیه‌ای، هنگام قرار داشتن در معرض شتاب ثابت 1 $\frac{m}{s^2}$ به مدت 10 ثانیه، باید بدانیم که سرعت خودرو در ابتدای این فاصله زمانی چه بوده است. به نحوی مشابه اگر بدانیم یک منبع ولتاژ ثابت 1 ولتی به مدت 10 ثانیه به مدار RC شکل 1-1 اعمال شده است، نمیتوانیم ولتاژ خازن را در انتهای این فاصله بیابیم، مگر اینکه ولتاژ اولیه آنرا بدانیم.
در حالت کلی برای حل یک معادله دیفرانسیل باید یک یا چند شرط کمکی مشخص کنیم، و در این صورت میتوانیم حداقل از لحاظ نظری، عبارت صریح خروجی را برحسب ورودی بیابیم. به بیانی دیگر، معادله دیفرانسیلی مثل معادله 2-95 قیدی بر رابطه ورودی و خروجی یک سیستم تحمیل میکند، ولی برای مشخص کردن کامل سیستم باید چند شرط کمکی نیز تعیین کنیم. به ازای شرایط کمکی متفاوت، روابط متفاوتی بین ورودی و خروجی بدست میاید. در اکثر بخشهای این کتاب تاکیدمام بر معادلات دیفرانسیل توصیف کننده سیستمهای LTI علی است. برای روشن کردن این موضوع و پرده برداشتن از خواص اصلی جوابهای معادلات دیفرانسیل، معادله 2-95 را به ازای یک سیگنال ورودی خاص حل میکنیم.

#### **مثال 2-14:**
<div dir="rtl">
حل معادله 2-95 را به ازای سیگنال ورودی زیر بیابید.
$$ x(t)=Ke^{3t}u(t)\;\;\;\;\;\;\;\;(96-2) $$
K یک عدد حقیقی ثایت است.
    جواب کامل معادله 2-96 مجموع یک جواب خصوصی $y_p(t)$ و یک جواب همگن $y_h(t)$ است؛ یعنی:
$$ y(t)=y_p(t)+y_h(t)\;\;\;\;\;\;\;\;(97-2) $$
جواب خصوصی معادله 2-95 را ارضا میکند و $y_h(t)$ جواب معادله دیفرانسیل همگن زیر است:
    $$\frac{dy(t)}{dt}+2y(t)=0\;\;\;\;\;\;\;\;(98-2)$$
یک روش معمول برای یافتن جواب خصوصی به ازای سیگنال ورودی نمایی معادله 2-96 این است که به دنبال جواب واداشته بگردیم، یعنی سیگنالی بصورت سیگنال ورودی. باتوجه به معادله 2-95 چون $x(t)=Ke^{3t}\;,t>0$ جواب زیر را برای $t>0$ پیشنهاد میکنیم:
$$y_p(t)=Ye^{3t}\;\;\;\;\;\;\;\;(99-2)$$
که در آن ؛ عددی است که باید تعیین شود. گذاشتن معادله 2-96 و 2-99 در معادله 2-95 بدست میدهد:
$$3Ye^{3t}+2Ye^{3t}=Ke^{3t}\;\;\;\;\;\;\;\;\;(100-2)$$
با حذف عامل مشترک $e^{3t}$ از دو طرف معادله 2-100 بدست میاوریم:
$$ 3Y+2Y=K\;\;\;\;\;\;\;\;(101-2)$$
یا
$$Y=\frac{K}{5}\;\;\;\;\;\;\;\;(102-2)$$
پس
$$y_p(t)=\frac{K}{5}e^{3t}\;\;\;,t>0\;\;\;\;\;\;\;\;(103-2)$$
برای تعیین $y_h(t)$ جوابی به شکل زیر پیشنهاد میکنیم:
$$y_h(t)=Ae^{st}\;\;\;\;\;\;\;\;\;(104-2)$$
و با گذاشتن آن در معادله 2-98 بدست میاوریم:
$$Ase^{st}+2Ae^{st}=Ae^{st}(s+2)=0\;\;\;\;\;\;\;\;\;(105-2)$$
باتوجه به این معادله باید داشته باشیم $s=-2$ و $Ae^{2t}$ به ازای هر مقدار A جواب معادله 2-98 است. با ترکیب معادلات 2-103 و 2-97، جواب معادله دیفرانسیل در $t>0$ بصورت زیر است:
$$y(t)=Ae^{2t}+\frac{K}{5}e^{3t}\;\;\;,t>0\;\;\;\;\;\;\;\;(106-2)$$
<div dir="rtl">
همانطور که قبلا گفتیم معادله دیفرانسیل 2-95 به تنهایی پاسخ $y(t)$ به ورودی $x(t)$ معادله 2-96 را بطور یکتا مشخص نمیکند. در واقع ثابت A معادله 2-106 هنوز مجهول است. برای تعیین مقدار A اید علاوه بر معادله دیفرانسیل 2-95 یک شرط کمکی نیز داشته باشیم. چنانچه در مسئله 2-34 خواهیم دید، به ازای شرایط کمکی مختلف جوابهای $y(t)$ متفاوت، و درنتیجه روابط ورودی-خروجی متفاوتی بدست میاید. همانطور که گفتیم در قسمت عمده‌ای از این کتاب تاکید بر معادلات دیفرانسیل و تفاضلی توصیف کننده سیستمهای LTI علی است، و در این حالت شرایط کمکی همان شرایط سکون ابتدایی هستند. یعنی همانطور که در مسئله 1-44 نشان داده شده برای سیستمهای LTI علی، اگر در $t<{t_0}$ داشته باشیم $x(t)=0$، آنگاه $y(t)$ نیز باید در $ t<{t_0}$ صفر باشد. باتوجه به معادله 2-96 میبینیم که در این مثال $x(t)$ در $t<0$ صفر است، پس شرط سکون ابتدایی مستلزم آن است که در $t<0$ داشته باشیم $y(t)=0$. معادله 2-106 را در $t=0$ حساب کرده و با قرار دادن $y(0)=0$ بدست میاوریم:
$$0=A+\frac{K}{5}$$
یا
$$A=-\frac{K}{5}$$
پس در t>0 
$$ y(t)=\frac{K}{5}[e^{2t}-e^{3t}]\;\;\;\;\;\;\;\;\;(107-2)$$
و در $y(t)=0,t<0$ که توسط شرط سکون ابتدایی تعیین میشود. با ترکیب این دو نتیجه جواب کامل زیر را میابیم:
$$y(t)=\frac{K}{5}[e^{2t}-e^{3t}]u(t)\;\;\;\;\;\;\;\;\;(108-2)$$
مثال 2-14 چند نکته بسیار مهم در مورد معادلات دیفرنسیل خطی با ضرائب ثابت، و سیستمهای نمایش داده شده با آنها را نشان میدهد. اول اینکه پاسخ به ورودی $x(t)$ در حالت کلی مجموع یک جواب خصوصی معادلات دیفرانسیل و یک جواب همگن، یعنی جواب معادله دیفرانسیل به ازای ورودی صفر، است. جواب همگن را غالبا پاسخ طبیعی سیستم مینامند. پاسخهای طبیعی مدارهای الکتریکی و سیستمهای مکانیکی ساده را طی مسائل 2-61 و 2-62 مورد بررسی قرار میدهیم.
در مثال 2-14 دیدیم که برای تعیین رابطه کامل بین ورودی و خروجی یک سیستم توصیف شده با معادله دیفرانسیلی چون معادله 2-95 باید شرایط کمکی معین کنیم. نتیجه ضمن این مطلب این است که همانطور که در مسئله 2-34 نشان داده میشود، انتخابهای مختلف شرایط کمکی به روابط متفاوتی بین ورودی و خروجی منجر میشود. در اکثر موارد شرایط سکون ابتدایی را برای سیستمهای توصیف شده با معادلات دیفرانسیل بکار میبریم. در این مثال چون در t<0 ورودی صفر بود، شرایط سکون ابتدایی شرط اولیه $y(0)=0$ را ایجاب میکرد. همانطور که گفتیم، و در مسئله 2-23 نشان میدهیم، با شرط سکون ابتدایی سیستم توصیف شده با معادله 2-95 LTI و علی است. برای مثال اگر ورودی معادله 2-96 را در 2 ضرب کنیم، خروجی حاصل دو برابر خروجی معادله 2-108 خواهد بود.
تاکید بر این نکته مهم است که شرط سکون ابتدایی، یک شرط اولیه صفر در یک زمان معین نیست؛ بلکه این زمان طوری تنظیم میشود که پاسخ تا زمان غیرصفر شدن ورودی صفر بماند. پس اگر برای سیستم LTI علی توصیف شده با معادله 2-95 در $t \leq t_0$ داشته باشیم $x(t)=0$، $y(t)$ در $t \leq t_0$ صفر است و برای یافتن جواب در $t>t_0$ باید شرط اولیه $y(t_0)=0$ را بکار برد. به عنوان یک مثال فیزیکی، بازهم مدار شکل 1-1 را که در مثال 1-8 مورد بحث قرار گرفت، در نظر بگیرید. سگون ابتدایی برای این مثال به این معنی است که تا وقتی منبع ولتاژ غیرصفری به این مدار اعمال نشود ولتاژ خازن صفر است. پس اگر ظهر امروز مدار را بکار گیریم، ولتاژ خازن تا ظهر امروز صفر است. همچنین اگر شروع کار در ظهر فردا باشد، ولتاژ اولیه خازن در ظهر فردا صفر خواهد بود.
این کثال همچنین به ما میگوید که چرا شرط سکون ابتدایی یک سیستم توصیف شده با معادله دیفرانسیل خطی با ضرائب ثابت، باعث تغییرناپذیری با زمان میشود. اگر بر روی این مدار با شروع از سکون سکون آزمایشی صورت دهیم، به فرض ثابت بودن ضرائب C و R با زمان، انتظار داریم نتایج آزمایش به دوشنبه یا سه شنبه بودن روز آزمایش بستگی نداشته باشد. یعنی اگر آزمایش مشابهی را در دو روز مختلف انجام دهیم و هر دو روز آزمایش را از سکون ابتدایی و در ساعت 12 شروع کنیم، جوابهایی یکسان بدست میاوریم، جوابهایی که نسبت بهم یک روز جابجایی دارند.
گرچه معادله دیفرانسیل مرتبه اول 2-95 را به عنوان محملی برای مباحث بالا بکار بردیم، ولی میتوانیم نتایج را به معادلات دیفرانسیل مرتبه بالا تعمیم دهیم. معادله دیفرانسیل خطی با ضرائب ثابت مرتبه N بصورت زیر است:
$$ \sum_{k=0}^{N}a_k\frac{d^ky(t)}{dt^k}=\sum_{k=0}^{M}b_k\frac{d^kx(t)}{dt^k}\;\;\;\;\;\;\;\;\;\;(109-2) $$
مرتبه معادله توسط بالاترین مشتق خروجی y(t) موجود در معادله تعیین میشود. در حالت $N=0$ معادله 2-109 بصورت زیر ساده میشود:
$$ y(t)=\frac{1}{a_0}\sum_{k=0}^{M}b_k\frac{d^kx(t)}{dt^k}\;\;\;\;\;\;\;\;\;\;(110-2) $$
در این حالت y(t) تابعی صریح از ورودی x(t) و مشتقهای ان است. به ازای $N\geq1$ معادله 2-109 خروجی را بطور ضمنی به ورودی مرتبط میکند. در اینصورت تحلیل معادله درست مانند بحث مربوط به معادله مرتبه اول 2-14 دنبال میشود. y(t) دو بخش دارد-جواب خصوصی معادله 2-109 به اضافه یک جواب معادله دیفرانسیل همگن زیر:
$$ \sum_{k=0}^{N}a_k\frac{d^ky(t)}{dt^k}=0\;\;\;\;\;\;\;\;\;\;(111-2) $$
جواب این معادله را پاسخ طبیعی سیستم مینامند.
مانند حالت مرتبه اول، معادله دیفرانسیل 2-109 خروجی را بطور کامل مشخص نمیکند و برای اینکه رابطه ورودی-خروجی سیستم بطور کامل معین شود به شرایط کمکی نیاز داریم. باز هم شرایط کمکی مختلف به روابط ورودی-خروجی متفاوت منجر میشوند، ولی در اغلب بخشهای کتاب هنگام کار با سیستمهای توصیف شده با معادلات دیفرانسیل شرایط سکون ابتدایی را بکار میبریم. یعنی اگر در $t\leq t_0$ داشته باشیم $x(t)=0$، فرض میکنیم در $t\leq t_0$ داریم $y(t)=0$ و بنابراین پاسخ در $t>t_0$ به کمک معادله دیفرانسیل 2-109 و شرایط کمکی زیر تعیین میشود:
$$ y(t_0)=\frac{dy(t_0)}{dt}=...=\frac{d^{N-1}y(t)}{dt^{N-1}}=0\;\;\;\;\;\;\;\;\;\;(112-2) $$
با شرط سکون ابتدایی سیستم توصیف شده با معادله 2-109 علی و LTI است. با داشتن شرایط اولیه بیان شده در معادله 2-112 میتوان به لحاظ نظری خروجی y(t) را با حل معادله دیفرانسیل، بصورت انجام شده در مثال 2-14 بدست آورد؛ چند مسئله آخر این فصل این روش را روشنتر میکنند. در فصلهای 4 و 9 ابزارهایی برای تحلیل سیستمهای LTI پیوسته در زمان پی میریزیم که حل معادلات دیفرانسیل را بسیار ساده میکنند و مخصوصا روشهای توانمندی برای تحلیل و مشخص کردن ویژگیهای سیستمهای توصیف شده با این معادلات را در اختیارمان میگذارند.
