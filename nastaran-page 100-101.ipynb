# خاصیت شرکت پذیری

یک خاصیت مهم و مفید دیگر کانولوشن خاصیت شرکت پذیری است. یعنی در حالت گسسته در زمان:

$$ x[n] * (h_1[n] * h_2[n]) = (x[n] * h_1[n]) * h_2[n] $$   (58-2)

و درحالت پیوسته زمان:

$$ x(t) * (h_1(t) * h_2(t)) = (x(t) * h_1(t)) * h_2(t) $$   (59-2)

با کار بر روی جمعها و انتگرالهای مربوط به این عبارتها میتوان آنهارا به سادگی ثابت کرد. مثالهای تاییدکننده این عبارتها در مسئله 2-43 داده شده است.
بخاطر وجود خاصیت شرکت پذیری عبارتهای زیر خالی از ابهام هستند:

$$ y[n] = x[n] * h_1[n] * h_2[n] $$   (60-2)

$$ y(t) = x(t) * h_1(t) * h_2(t) $$   (61-2)

به عبارت دیگر، طبق معادلات (2-58) و (2-59) فرقی نمیکند که کانولوشن این سیگنالها به چه ترتیبی انجام شود.
شکلهای 2-25 (الف) و (ب) تعبیری از خاصیت شرکت پذیری را برای سیستمهای گسسته در زمان نشان میدهند. در شکل 2-25 (الف) داریم:

$$ y[n] = w[n] * h_2[n] $$
$$ = (x[n] * h_1[n]) * h_2[n] $$

در شکل 2-25 (ب) داریم:

$$ y[n] = x[n] * h[n] $$
$$ = x[n] * (h_1[n] * h_2[n]) $$

اتصال سری دو سیستم شکل 2-25 (الف)، طبق خاصیت شرکت پذیری ، معادل سیستم شکل 2-25 (ب) است. این امر را میتوان به اتصال سری هر تعدادی سیستم خطی تغییرناپذیر با زمان تعمیم داد و برای حالت پیوسته در زمان  نیز تعبیر و نتیجه مشابهی برقرارست.
با استفاده از خاصیت شرکت پذیری و جابجایی کانولوشن خاصیت مهم دیگری برای سیستمهای خطی تغییرناپذیر با زمان به دست می‌آوریم. شکلهای 2-25 (الف) و (ب) نشان میدهند که پاسخ ضربه دو سیستم سری، با کانولوشن پاسخ ضربه آن دو سیستم برابر است. چون کانولوشن خاصیت جابه‌جایی دارد، کانولوشن \(h_1[n]\) و \(h_2[n]\) را میتوان به هر ترتیبی انجام داد. پس شکلهای 2-25 (ب) و (ج) هم ارزند و طبق خاصیت شرکت پذیری، این سیستم‌ها با سیستم شکل 2-25 (و) نیز که همانند شکل 2-25 (الف) ترکیب سری دو سیستم است، هم ارزند. پس پاسخ ضربه ترکیب متوالی دو سیستم خطی تغییر ناپذیر با زمان، به ترتیب سیستمها بستگی ندارد.
در واقع این امر برای هر تعداد سیستم متوالی شده‌ای درست است: تا وقتی به پاسخ ضربه سیستم کل علاقمندیم، ترتیب اتصال سیستمها مهم نیست. این خاصیت برای حالت پیوسته در زمان نیز برقرار است.
تاکید بر این نکته مهم است که رفتار سیستمهای خطی تغییرناپذیر با زمان سری شده، و خصوصا استقلال رفتار سیستم کل از ترتیب سیستمهای سری منحصر به این سیستمهاست. برعکس در حالت کلی نمیتوان جای سیستمهای غیرخطی را عوض کرد. زیرا چنین کاری باعث تغییر پاسخ کل سیستم میشود.
برای مثال دو سیستم بی حافظه را در نظر بگیرید که یکی ورودی را در 2 ضرب میکند و دیگری ورودی را مربعی میکند. اگر ابتدا ضرب و سپس مربع کنیم داریم:

$$ y[n] = 4x[n]^2 $$

اگر عمل ضرب بعد از به توان رساندن صورت گیرد، داریم:

$$ y[n] = 2x[n]^2 $$

پس امکان جابه‌جا کردن ترتیب سیستمهای سری، منحصر به سیستمهای خطی تغییرناپذیر با زمان است. در واقع چنان که در مسئله 2-51 نشان خواهیم داد، برای برقراری خاصیت فوق در حالت کلی، هم خطی بودن لازم است و هم تغییرناپذیری با زمان
